### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda['Tienda']='Tienda1'
tienda2 = pd.read_csv(url2)
tienda2['Tienda']='Tienda2'
tienda3 = pd.read_csv(url3)
tienda3['Tienda']='Tienda3'
tienda4 = pd.read_csv(url4)
tienda4['Tienda']='Tienda4'

df=pd.concat([tienda,tienda2,tienda3,tienda4])
df.head()
```

#1. Análisis de facturación



```python
ventas = df.groupby('Tienda')['Precio'].sum()
ventas_formateadas = ventas.apply(lambda x: f"${x:,.0f}")
print(ventas_formateadas)
```

# 2. Ventas por categoría

```python
ventas_categoria = df.groupby('Categoría del Producto')['Precio'].sum()
ventas_categoria_formateada = ventas_categoria.apply(lambda x: f"${x:,.0f}")
print(ventas_categoria_formateada)
```

# 3. Calificación promedio de la tienda


```python
calificacion_promedio=df.groupby('Tienda')['Calificación'].mean()
calificacion_promedio_formateada = calificacion_promedio.apply(lambda x: f"{x:.1f}")
print(calificacion_promedio_formateada)
```

# 4. Productos más y menos vendidos

```python
productos_mas_vendidos = df['Producto'].value_counts().sort_values(ascending=False)


print("Producto más vendidos:", productos_mas_vendidos.head(3))
print("Producto menos vendido:", productos_mas_vendidos.tail(3))
```

# 5. Envío promedio por tienda

```python
envio_promdeio=df.groupby('Tienda')['Costo de envío'].mean()
envio_promedio_formateado = envio_promdeio.apply(lambda x: f"${x:,.0f}")
print(envio_promedio_formateado)
```
